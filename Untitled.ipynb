```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import pandas as pd
import random
import statsmodels.api as sm
```

```python
R=250
beta0=[]
beta1=[]
for i in range(R):
    x=np.random.randn(100)+5
    e=stats.norm(0, 1).rvs(100)
    y=10+3*x+e
    X=sm.add_constant(x)
    mod=sm.OLS(y, X).fit()
    beta0.append(mod.params[0])
    beta1.append(mod.params[1])
print("Mean of the estimated \beta1 is ", sum(beta1)/len(beta1))
plt.hist(beta1, label="Mean of the distribution {:3.2f}".format(sum(beta1)/len(beta1)))
plt.title("Figure 4: Plot of values of unbiased slope coefficient estimates")
plt.legend()
plt.show()
```

```python
R=250
sample_size=[10, 20, 1000]
beta0=[]
beta1=[]
ff=6
for samplesize in sample_size:
    for i in range(R):
        x=np.random.randn(100)+5
        e=stats.norm(0, 1).rvs(100)+0.25*x
        y=10+3*x+e
        X=sm.add_constant(x)
        mod=sm.OLS(y, X).fit()
        beta0.append(mod.params[0])
        beta1.append(mod.params[1])
    print("Mean of the estimated \beta1 is: ", sum(beta1)/len(beta1), "with sample size: ", samplesize)
    plt.hist(beta1, label="Mean {:3.2f} \nSample size {:3.2f}".format(sum(beta1)/len(beta1), samplesize))
    plt.title("Figure {:3.0f}: Biased estimaes of coefficient and their distribution".format(ff))
    plt.legend()
    plt.show()
    ff=ff+1
```
